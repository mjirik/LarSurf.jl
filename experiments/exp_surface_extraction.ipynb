## Statistics

```python
# dfs[dfs["fcn"]=="LarSurf Parallel"]["Time [s]"]
```

```python
from scipy import stats
# https://stackoverflow.com/questions/15984221/how-to-perform-two-sample-one-tailed-t-test-with-numpy-scipy/15984310

alpha=0.01
st = stats.ttest_ind(dfs[dfs["fcn"]=="LarSurf Parallel"]["Time [s]"], dfs[dfs["fcn"]=="Marching Cubes"]["Time [s]"])
display(st)
# stats.tt

print("pokud je pvalue mensi nez zvolena hladina vyznamnosti (0.01=1%), je vsechno ok") 
#mozna staci i dvojnasobek hladiny vyzamnosi
print("statistic musi byt vetsi nez 0")
print()
print("H0: A >= B               H1(alternative): B > A")
print("ttest_ind(B, A)")
print(f"p/2 < alpha     =>       {st.pvalue} < {alpha}      =>      {st.pvalue/2 < alpha}")
print("If true                  you can reject H0 ")
print("If false                 you cannot reject H0 ")
# if  (st.pvalue/2 < alpha ) & (st.pvalue > 0)
#     print()

dtt.save(st.pvalue, "ircad_pvalue")
dtt.save(st.statistic, "ircad_statistics")
```
